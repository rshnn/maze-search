# Generating Hard Mazes

# Trivial Case

The trivial solution to this problem is generating N number of random mazes, and returning the one that is the most difficult according to some metric.  The algorithms we will use in these cases in A* using Manhattan distance and DFS.  From part 2, we saw that A* was among the best performers and DFS was also able to very efficiently solve mazes.  

```python
N = 1000 
dim = 20 
p = 0.2 

# inputs for search algos 
inital_state = Position(0,0)
goal_state = Position(dim-1, dim-1)

```

## Trivial Using A* Manhattan distance 

```python
(longest_path, nodes_expnded, fringe, _) = SearchHard.trivial(N, dim, p, search_algo=Search.A_star_man)
```

```python
# Max fringe 

m = fringe[1].max_fringe_size

_, path, _ = Search.A_star_man(fringe[0], inital_state, goal_state)
f = fringe[0].draw_path(path, plot_title="Trivial A* - Largest Max Fringe Size - {} nodes".format(m))


# Longest Path 
m = longest_path[1].path_length

_, path, _ = Search.A_star_man(longest_path[0], inital_state, goal_state)
f = longest_path[0].draw_path(path, plot_title="Trivial A* - Longest Path - {} units".format(m))



# Total Nodes Expanded 
m = nodes_expnded[1].total_nodes_expanded

_, path, _ = Search.A_star_man(nodes_expnded[0], inital_state, goal_state)
f = nodes_expnded[0].draw_path(path, plot_title="Trivial A* - Most Nodes Expanded - {} nodes".format(m))
```

## Trivial Using DFS


```python
N = 1000 
dim = 20 
p = 0.2 

# inputs for search algos 
inital_state = Position(0,0)
goal_state = Position(dim-1, dim-1)

(longest_path_dfs, nodes_expnded_dfs, fringe_dfs, _) = SearchHard.trivial(N, dim, p, search_algo=Search.dfs)
```

```python
# Largest Fringe 
m = fringe_dfs[1].max_fringe_size

_, p, _ = Search.dfs(fringe_dfs[0], inital_state, goal_state)
f = fringe_dfs[0].draw_path(p, plot_title="Trivial DFS - Largest Max Fringe Size - {} nodes".format(m))


# Longest Path 
m = longest_path_dfs[1].path_length

_, p, _ = Search.dfs(longest_path_dfs[0], inital_state, goal_state)
f = longest_path_dfs[0].draw_path(p, plot_title="Trivial DFS - Longest Path - {} units".format(m))

# Most nodes expanded 
m = nodes_expnded_dfs[1].total_nodes_expanded

_, p, _ = Search.dfs(nodes_expnded_dfs[0], inital_state, goal_state)
f = nodes_expnded_dfs[0].draw_path(p, plot_title="Trivial DFS - Most Nodes Expanded - {} nodes".format(m))
```
