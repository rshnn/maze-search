# 2a 
- Find a map size (dim) that is large enough to produce maps that require some work to solve, but small enough
that you can run each algorithm multiple times for a range of possible p values. 
- How did you pick a dim?

### Answer
- dim=100



```python
p_val=0.3 

dim_array=np.linspace(50,150,101) #using dim
initial_state = Position(0,0)


euc_average_nodes_expanded=[]
euc_average_fringe_size=[]
man_average_nodes_expanded=[]
man_average_fringe_size=[]
bfs_average_nodes_expanded=[]
bfs_average_fringe_size=[]
dfs_average_nodes_expanded=[]
dfs_average_fringe_size=[]
bi_bfs_average_nodes_expanded=[]
bi_bfs_average_fringe_size=[]


num_random_mazes=10

for dim in dim_array:
    print(dim)
    euc_nodes_expanded=[]
    euc_fringe_size=[]
    man_nodes_expanded=[]
    man_fringe_size=[]
    dfs_nodes_expanded=[]
    dfs_fringe_size=[]
    bfs_nodes_expanded=[]
    bfs_fringe_size=[]
    bi_bfs_nodes_expanded=[]
    bi_bfs_fringe_size=[]
    
    for i in range(0,num_random_mazes): #generate x random mazes

        maze = Maze(int(dim), p_val, 0)
        goal_state = Position(maze.dim-1, maze.dim-1)
        #perform all algorithms
        euc_success, euc_path, euc_metrics = Search.A_star_euc(maze, initial_state, goal_state, question='2a')
        man_success, man_path, man_metrics = Search.A_star_man(maze, initial_state, goal_state)
        bfs_success, bfs_path, bfs_metrics = Search.bfs(maze, initial_state, goal_state)
        dfs_success, dfs_path, dfs_metrics = Search.dfs(maze, initial_state, goal_state)
        bi_bfs_success, bi_bfs_path, bi_bfs_metrics = Search.bid_bfs(maze, initial_state, goal_state,question='2a')
        

        if euc_success:
               #if they are successful, determine the total nodes expanded and the maximum fringe size
            euc_nodes_expanded.append(euc_metrics.total_nodes_expanded)
            euc_fringe_size.append(euc_metrics.max_fringe_size)
            man_nodes_expanded.append(man_metrics.total_nodes_expanded)
            man_fringe_size.append(man_metrics.max_fringe_size)
            bfs_nodes_expanded.append(bfs_metrics.total_nodes_expanded)
            bfs_fringe_size.append(bfs_metrics.max_fringe_size)
            dfs_nodes_expanded.append(dfs_metrics.total_nodes_expanded)
            dfs_fringe_size.append(dfs_metrics.max_fringe_size)
            bi_bfs_nodes_expanded.append(bi_bfs_metrics.total_nodes_expanded)
            bi_bfs_fringe_size.append(bi_bfs_metrics.max_fringe_size)
            
        
    #calculate the average nodes expanded and the average max fringe size per p value
    euc_average_nodes_expanded.append(statistics.mean(euc_nodes_expanded))
    euc_average_fringe_size.append(statistics.mean(euc_fringe_size))
    man_average_nodes_expanded.append(statistics.mean(man_nodes_expanded))
    man_average_fringe_size.append(statistics.mean(man_fringe_size))
    bfs_average_nodes_expanded.append(statistics.mean(bfs_nodes_expanded))
    bfs_average_fringe_size.append(statistics.mean(bfs_fringe_size))
    dfs_average_nodes_expanded.append(statistics.mean(dfs_nodes_expanded))
    dfs_average_fringe_size.append(statistics.mean(dfs_fringe_size))
    bi_bfs_average_nodes_expanded.append(statistics.mean(bi_bfs_nodes_expanded))
    bi_bfs_average_fringe_size.append(statistics.mean(bi_bfs_fringe_size))
    
    #plot these metrics for comparison
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.plot(dim_array, euc_average_nodes_expanded, dim_array, man_average_nodes_expanded, dim_array, bfs_average_nodes_expanded, dim_array, dfs_average_nodes_expanded,dim_array, bi_bfs_average_nodes_expanded)
plt.legend(["Euclidean Distance","Manhattan Distance", "Breadth First Search","Depth First Search","Bi-Directional BFS"])
plt.xlabel("dim")
plt.ylabel("Average number of nodes expanded")
plt.title("Average nodes expanded comparison")

plt.subplot(133)
plt.plot(dim_array,euc_average_fringe_size, dim_array, man_average_fringe_size, dim_array,bfs_average_fringe_size, dim_array, dfs_average_fringe_size,dim_array, bi_bfs_average_fringe_size)
plt.legend(["Euclidean Distance","Manhattan Distance", "Breadth First Search","Depth First Search","Bi-Directional BFS"])
plt.xlabel("dim")
plt.ylabel("Average max fringe size")
plt.title("Max fringe size comparison")
```

## 2c
- Given dim, how does maze-solvability depend on p? 
- For a range of p values, estimate the probability that a maze will be solvable by generating multiple mazes and checking them for solvability. 
- What is the best algorithm to use here? 
- Plot density vs solvability, and try to identify as accurately as you can the threshold p0where for p < p0, most mazes are solvable, but p > p0, most mazes are not solvable.

```python
p=np.linspace(0.01,0.99,20) #generate 20 evenly spaces p values from 0.01 to 0.99
dim=100

num_random_mazes=100

initial_state = Position(0,0)
goal_state = Position(dim-1, dim-1)

range_of_p_successes=[]

for p_val in p: #for each pvalue
    
    num_successes=0
    
    for i in range(0,num_random_mazes): #generate x number of random mazes
        maze = Maze(dim, p_val, 0)
        
        success, path, metrics = Search.A_star_euc(maze, initial_state, goal_state, question='2c')
        
        if success: #if the maze is solvable, add it to the count
            num_successes=num_successes+1
            
    #calculate the average success rate
    range_of_p_successes.append(num_successes/num_random_mazes)
    

#plot the average success rate as p varies
plt.plot(p, range_of_p_successes)     
plt.title('Solvability vs. P value (density)')
plt.xlabel('P value')
plt.ylabel('Percentage of mazes solvable')
    
```

### Improvement Analysis

```python
p0=0.3
p=np.linspace(0.01,p0,10)
dim=100
initial_state = Position(0,0)
goal_state = Position(dim-1, dim-1)

dfs_average_nodes_expanded=[]
dfs_average_fringe_size=[]
se_dfs_average_nodes_expanded=[]
se_dfs_average_fringe_size=[]
dfs_average_length=[]
se_average_length=[]

num_random_mazes=50

for p_val in p: #for p in a range of p values from 0.01 to p0
    
    dfs_nodes_expanded=[]
    dfs_fringe_size=[]
    se_dfs_nodes_expanded=[]
    se_dfs_fringe_size=[]
    dfs_length=[]
    se_length=[]
    
    for i in range(0,num_random_mazes): #generate x random mazes
        
        maze = Maze(dim, p_val, 0)
        
        #perform both regular DFS and Southeast DFS on them
        dfs_success, dfs_path, dfs_metrics = Search.dfs(maze, initial_state, goal_state)
        se_dfs_success, se_dfs_path, se_dfs_metrics = Search.dfs_southeast(maze, initial_state, goal_state)
        
        if dfs_success:
            #if a path is found, add the total nodes expanded and path length and max fringe size of the algorithms
            dfs_nodes_expanded.append(dfs_metrics.total_nodes_expanded)
            dfs_fringe_size.append(dfs_metrics.max_fringe_size)
            se_dfs_nodes_expanded.append(se_dfs_metrics.total_nodes_expanded)
            se_dfs_fringe_size.append(se_dfs_metrics.max_fringe_size)
            dfs_length.append(len(dfs_path))
            se_length.append(len(se_dfs_path))
            
    print(p_val)
            
            
    #compute the average max fringe size and the average number of nodes expanded for each aglorithm for each p
    dfs_average_nodes_expanded.append(statistics.mean(dfs_nodes_expanded))
    dfs_average_fringe_size.append(statistics.mean(dfs_fringe_size))
    se_dfs_average_nodes_expanded.append(statistics.mean(se_dfs_nodes_expanded))
    se_dfs_average_fringe_size.append(statistics.mean(se_dfs_fringe_size))
    dfs_average_length.append(statistics.mean(dfs_length))
    se_average_length.append(statistics.mean(se_length))
print('DONE')
```

```python
#plot the data
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.plot(p, dfs_average_nodes_expanded, p, se_dfs_average_nodes_expanded)
plt.legend(["Regualr DFS","SouthEast DFS"])
plt.xlabel("P value")
plt.ylabel("Average number of nodes expanded")
plt.title("Average nodes expanded comparison")

plt.subplot(132)
plt.plot(p,dfs_average_fringe_size, p, se_dfs_average_fringe_size)
plt.legend(["Regualr DFS","SouthEast DFS"])
plt.xlabel("P value")
plt.ylabel("Average max fringe size")
plt.title("Max fringe size comparison")

plt.subplot(133)
plt.plot(p,dfs_average_length, p, se_average_length)
plt.legend(["Regualr DFS","SouthEast DFS"])
plt.xlabel("P value")
plt.ylabel("Averange path length")
plt.title("Averange path length comparison")
```
